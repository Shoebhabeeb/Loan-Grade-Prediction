## Question 3 [5 Pts]

How does a PCA work [Principal Component Analysis] ? 

### Solution

Steps of working of PCA – 
a.	Calculate the covariance matrix X of data points.
b.	Calculate the eigen vectors and corresponding eigen values. 
c.	Sort the eigen vectors according to their eigen values in decreasing order.
d.	Choose first k eigen vectors which will be the new k dimensions.
e.	Transform the original n dimensional data points into k dimensions.

## Question 6 [10 Pts]

What is clustering? Explain how K-Means Clustering Algorithm works?

### Solution

Clustering in data mining is an unsupervised learning procedure whose goal is to find homogenous subgroups among the observations. A cluster of data objects can be considered as one group. Advantage of clustering over classification is that it is adaptable to changes and helps group useful features that distinguish different groups. 

Clustering analysis is used in applications such as market research, pattern recognition, image processing and data analysis. As a data mining function, clustering serves as a tool to gain insight into distribution of data to observe characteristics of each cluster.

Working of K-means clustering:-

Objective of K-means clustering is to group similar data points and discover underlying patterns. K is the number of clusters and must be fixed in advance i.e. the number of centroids in the dataset. 
i.	Randomly assign each sample to cluster from 1..K, until all the data points are assigned to a cluster.
ii.	Find centroid for each cluster i.e. the average of all samples in the cluster.
iii.	Reassign each sample to the nearest centroid.
iv.	The algorithm converges to local minimum of 
 
The K-means clustering algorithm halts creating and optimizing clusters when either –
a.	The centroids have stabilized i.e. no change in centroids values because the clustering is successful, or,
b.	The defined number of iterations is achieved.
K-means clustering can be achieved programmatically using python’s Scikit-learn library.

### Solution

K-Means Advantages & Disadvantages:
Advantages-
i.	Easy to implement.
ii.	With large number of variables, K-Means may be computationally faster than hierarchical clustering, if k is small.
iii.	K-Means may produce tighter clusters than hierarchical clustering.

i.	Difficult to predict the number of clusters i.e. the k value.
ii.	Initial seeds have strong impact on results.
iii.	Order of the data has impact on the final results.
iv.	Rescaling the datasets by normalization or standardization completely changes the results.

Hierarchical clustering Advantages & Disadvantages:
Advantages-
i.	Hierarchical clustering outputs a hierarchy, i.e. structure that is more informative than unstructured set of clusters returned by k-means. Hence, it is easy to decide the number of clusters by looking at the dendrogram.
ii.	Easy to implement.

Disadvantages-
i.	It is not possible to undo a step already executed, i.e. once the instances have been assigned to a cluster, it can not be moved to another cluster.
ii.	Time complexity: It is not suitable for huge datasets.
iii.	Very sensitive to outliers.

### Relevant Information:

   The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, 
   in order to access if the product (bank term deposit) would be (or not) subscribed. 

   There are two datasets:

      2) Number of Attributes: 16 + output attribute.

### Attribute information:

   Input variables:
   #### bank client data:
   
   1 - age (numeric)
   
   2 - job : type of job (categorical: 
   "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
                                       "blue-collar","self-employed","retired","technician","services") 
                                       
   3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)
   
   4 - education (categorical: "unknown","secondary","primary","tertiary")
   
   5 - default: has credit in default? (binary: "yes","no")
   
   6 - balance: average yearly balance, in euros (numeric) 
   
   7 - housing: has housing loan? (binary: "yes","no")
   
   8 - loan: has personal loan? (binary: "yes","no")
   
   #### related with the last contact of the current campaign:
   
   9 - contact: contact communication type (categorical: "unknown","telephone","cellular") 
   
  10 - day: last contact day of the month (numeric)
  
  11 - month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")
  
  12 - duration: last contact duration, in seconds (numeric)
  
   #### other attributes:
   
  13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
  
  14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)
  
  15 - previous: number of contacts performed before this campaign and for this client (numeric)
  
  16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")
  


  Output variable (desired target):
  
  17 - y - has the client subscribed a term deposit? (binary: "yes","no")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# Loading the csv file data
df = pd.read_csv('homework_2.csv')

#Formatting dataframe splitting by ';'
formattedDF = pd.DataFrame(columns=['age','job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome','y'])
formattedDF[['age','job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome','y']] = df['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(';', expand=True)

# Data Preprocessing
ohe = OneHotEncoder(sparse=False)
formattedDF_transformed = ohe.fit_transform(formattedDF)

# Creating PCA instance: pca
pca = PCA(n_components=17) # for 17 columns
principalComponents = pca.fit_transform(formattedDF_transformed)

print("Eigen vector values::")
print(pca.components_)

print("\n Variance in each dimension::")
print(pca.explained_variance_)

#Plot explained variances
features = range(pca.n_components_)
plt.bar(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_)
plt.tight_layout()
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

The bar plot above displays PCA features vs Variance % of the 17 components of the given csv dataset.
It can be observed that there is a variance drop-off point from feature 0 to feature 1 by 0.01% approximately. Also, a significant variance drop-off from feature 2 to feature 3 by 0.02% approximately.
Another minor variance drop-off can be observed from feature 5 to 6, while rest of the feature's variance drop decrements gradually upto 0.01.

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='green')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar()
```

From the above scatter plot, we can observe that there are 3 clearly distinguishable clusters formed.

```python
ks = range(1, 17)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(PCA_components.iloc[:,:3])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='blue')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

The above plot shows a slow decrease in intertia after 3 clusters i.e after k=3. The value of the inertia is no longer significant after 3rd cluster, neither is the variance of the rest of the data after the elbow point. Hence, everything after k=3 can be discarded for further analysis.

Inferences from PCA algorithm output and K-means clustering output:
On applying K-Means and PCA algorithm on given bank marketing dataset, it can be observed that the number of important features which drives the analysis of the data is 3. Features Age, Job and marital status are the driving features for the bank marketing fraud prediction. Other attributes like campus, p-days, pnumber and outcome doesn't really matter for analysis of fraud predictions from the dataset. Features like default, balance and loan can help in the analysis to some extent.
